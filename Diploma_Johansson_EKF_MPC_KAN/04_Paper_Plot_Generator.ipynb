```python
import matplotlib.pyplot as plt
import pandas as pd
import scienceplots  # This registers the styles
import numpy as np

# Define your HIL time step
dt_hil = 0.003  # 3ms

# 1. Load Data
df_mpc = pd.read_csv('Data/DisturbanceHandling_MPC.csv')
df_kan = pd.read_csv('Data/DisturbanceHandling_KAN.csv')

# Create the time vector based on the number of rows
df_mpc['time'] = np.arange(len(df_mpc)) * dt_hil
df_kan['time'] = np.arange(len(df_kan)) * dt_hil

# 2. Set the Style
plt.style.use(['science', 'ieee'])

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Times"],
})

# 3. Create Figure
# IEEE Column width is ~3.5 inches.
# Aspect ratio 4:3 or 16:9 depending on vertical space.
fig, ax = plt.subplots(figsize=(3.5, 2.5))

# 4. Plotting (The "Money Shot" Disturbance Graph)
# Use thin lines (linewidth=0.8 - 1.0 is standard for IEEE)
ax.plot(df_mpc['time'], df_mpc['h1'], label='MPC', linestyle='-', color='black', linewidth=1)
ax.plot(df_kan['time'], df_kan['h1'], label='Symbolic KAN', linestyle='--', color='red', linewidth=1)

# Target Line
ax.axhline(10.0, color='gray', linestyle=':', linewidth=0.8, label='Reference')

# 5. Labeling
ax.set_xlabel('Time (s)')
ax.set_ylabel('Water Level $h_1$ (cm)') # Use LaTeX syntax for math
ax.legend(fontsize=7, frameon=True, loc='lower right') # Small font for legend
ax.set_xlim([20, 50]) # Zoom in on the disturbance
ax.set_ylim([8, 14])

# 6. Saving
# ALWAYS save as .pdf or .eps for LaTeX.
plt.savefig('Fig4_Disturbance_Rejection.pdf', dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load Data
df_mpc = pd.read_csv('Data/HIL_Data_CEMPC_NMP_STM32.csv')
df_kan = pd.read_csv('Data/HIL_Data_KAN_NMP_STM32.csv')   

# Define Target
TARGET = 10.0

def calculate_metrics(df, start_time=0):
    
    # Assuming 'time' column exists or generating it
    dt = 0.003
    if 'time' not in df.columns:
        df['time'] = df.index * dt
        
    data_slice = df[df['time'] >= start_time]
    
    # Calculate Error Vectors
    e1 = data_slice['h1'] - TARGET
    e2 = data_slice['h2'] - TARGET
    
    # 1. RMSE (Root Mean Square Error) - The "Standard"
    rmse_h1 = np.sqrt(np.mean(e1**2))
    rmse_h2 = np.sqrt(np.mean(e2**2))
    total_rmse = np.sqrt(np.mean(e1**2 + e2**2)) # Combined
    
    # 2. IAE (Integral Absolute Error) - Good for accumulated deviation
    # Sum of absolute error * dt
    iae_h1 = np.sum(np.abs(e1)) * dt
    iae_h2 = np.sum(np.abs(e2)) * dt
    
    return total_rmse, rmse_h1, rmse_h2

print("--- TRACKING PERFORMANCE (0 - 50s) ---")
mpc_rmse, _, _ = calculate_metrics(df_mpc, start_time=0)
kan_rmse, _, _ = calculate_metrics(df_kan, start_time=0)
print(f"MPC Total RMSE: {mpc_rmse:.4f} cm")
print(f"KAN Total RMSE: {kan_rmse:.4f} cm")

print("\n--- STEADY STATE PERFORMANCE (40s - 50s) ---")
mpc_ss, _, _ = calculate_metrics(df_mpc, start_time=40)
kan_ss, _, _ = calculate_metrics(df_kan, start_time=40)
print(f"MPC Steady-State RMSE: {mpc_ss:.4f} cm")
print(f"KAN Steady-State RMSE: {kan_ss:.4f} cm")
```
